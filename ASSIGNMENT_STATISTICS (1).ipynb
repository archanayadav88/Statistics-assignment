**Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.**

**Answer**:
#Descriptive Statistics:-
**Definition**:

Descriptive statistics is used to summarize, organize, and describe the main features of a dataset.

It deals with what the data shows — without making any predictions or generalizations beyond the data collected.

**Purpose**:
To present data in a meaningful way so patterns can be seen easily.

Common Tools Used:

Mean, Median, Mode (measures of central tendency)

Range, Variance, Standard Deviation (measures of spread)

Tables, Graphs, Charts (visual representation)

**Example**:
Suppose you collected the marks of 5 students in a test:
70, 80, 90, 60, 100

Descriptive statistics would include:

Mean (Average): (70 + 80 + 90 + 60 + 100) / 5 = 80

Median: 80

Range: 100 - 60 = 40

#Inferential Statistics
**Definition**:

Inferential statistics is used to make predictions, inferences, or generalizations about a population based on a sample of data.

It goes beyond the data to make conclusions about a larger group.

**Purpose**:
To draw conclusions or test hypotheses about a population using sample data.

Common Tools Used:

Hypothesis Testing

Confidence Intervals

Regression Analysis

**Example**:Suppose you take a sample of 100 students from a school to find the average height.
The sample mean height is 150 cm.

"The average height of all students in the school is likely around 150 cm, with a 95% confidence interval of 147-153 cm."

**In Simple Terms:**
1)Descriptive statistics → "What does my data show?"

2)Inferential statistics → "What can I conclude or predict from my data?"



**Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.**

**Answer**:
**Definition**:Sampling in statistics is the process of selecting a small group (sample)from a larger group(population)to collect data and make conclusions about the entire population.

Since studying an entire population is often time-consuming, costly, and impractical, we study a representative portion instead.

**Example**:

Suppose there are 10,000 students in a university, and you want to know the average height of students.
Instead of measuring everyone, you might select 200 students randomly — that group of 200 is your sample.
From this sample, you can estimate the average height for all 10,000 students.

#Difference between Random Sampling and Stratified Sampling:-

**Random Sampling**

#Definition:
In random sampling, every individual in the population has an equal chance of being selected.

#Purpose:
-To avoid bias and ensure that the sample represents the population fairly.

#Example:
You have a list of 1,000 students and you use a computer-generated random number to pick 100 of them for your survey.
→ Each student has an equal chance of being chosen.

#Advantages:
1)Simple and easy to understand.
2)Minimizes selection bias.

#Disadvantages:
-May not represent all subgroups equally (e.g., boys and girls, departments, etc.).

**Stratified Sampling**

#Definition:
In stratified sampling, the population is divided into subgroups (strata) based on certain characteristics (like age, gender, class, or income), and then random samples are taken from each stratum.

#Purpose:
-To ensure that all key subgroups are represented in the sample.

#Example:
In a university with 60% girls and 40% boys, you can divide students into two strata (boys and girls).
Then, randomly select 60 girls and 40 boys out of 100 students to maintain the same ratio as the population.

#Advantages:
1)Ensures representation of all subgroups.
2)Provides more accurate and reliable results.

#Disadvantages:
-More complex and time-consuming to organize.

**In Simple Terms**

1)Random Sampling → Pick anyone randomly.

2)Stratified Sampling → Divide into groups, then pick randomly from each group.


**Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.**

**Answer**:
In statistics,mean, median, and mode are the three main measures of central tendency.
They help us understand the center or average value of a dataset.

**Mean(Average)**

**Definition**:
The mean is the sum of all values divided by the total number of values.
It represents the arithmetic average of the data.

**Formula**:

Mean=Number of observations/Sum of all observations
	​
**Example:**
If the marks of 5 students are 10, 20, 30, 40, and 50,

Mean=10+20+30+40+50/5

=150/5=30

So, the mean = 30.


**Median**

**Definition**:
The median is the middle value when all the data items are arranged in ascending or descending order.
If there are two middle values, the median is their average.

**Example**:
Data: 10, 20, 30, 40, 50
→ The middle value is 30, so the median = 30.

If data is 10, 20, 30, 40, 50, 60
→ The two middle values are 30 and 40

Median=30+40/2
=35


**Mode**

**Definition**:
The mode is the value that occurs most frequently in a dataset.
A set of data may have one mode, more than one mode, or no mode at all.

**Example**:
Data: 5, 8, 8, 9, 10, 8, 7
→ The most frequent value is 8, so mode = 8.


**Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data? **

**Answer:**

**1:-Skewness:-**

**Definition**:
Skewness measures the asymmetry (lack of symmetry) of a data distribution.
It tells us which side of the mean (left or right) the data are more stretched out.

Symmetrical distribution: Skewness = 0
→ Mean = Median = Mode (e.g., Normal distribution)

-Positively skewed (Right-skewed):
# Skewness > 0

1)The tail is longer on the right side.
2)Most data values are concentrated on the left, but there are a few large values (outliers) on the right.

**Example**: Income distribution (most people earn average or below average income, a few earn very high).

-Negatively skewed (Left-skewed):
#Skewness < 0

1)The tail is longer on the left side.
2)Most data values are concentrated on the right, but there are a few small values (outliers) on the left.

**Example**: Age at retirement (most retire around 60–65, few much earlier).


**2:-Kurtosis**:-

**Definition**:
Kurtosis measures the “tailedness” or peakedness of a data distribution — how heavy or light the tails are compared to a normal distribution.

-Mesokurtic (Kurtosis ≈ 3)
→ Normal distribution (average tails and peak).

-Leptokurtic (Kurtosis > 3)
→ Peaked in the center and heavy tails — more outliers than normal.
→ Example: Data with many extreme values.

-Platykurtic (Kurtosis < 3)
→ Flat-topped distribution and light tails — fewer outliers than normal.

(Sometimes “excess kurtosis” is used, which is simply kurtosis-3, so the normal distribution has excess kurtosis = 0.)

**What a Positive Skew Implies**
-If the skewness is positive:

1)The mean > median > mode

2)The right tail (higher values) is longer

3)There are a few large outliers

4)Most data points are clustered toward lower values

**Example**:
If exam marks are positively skewed, it means most students scored low to average marks, but a few scored very high marks, stretching the tail to the right.


```python
#Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

#solution:-
import statistics as stats

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = stats.mean(numbers)
median_value = stats.median(numbers)
mode_value = stats.mode(numbers)

print("Numbers:", numbers)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)


#Explanation:-

#Mean (20.0): Average of all numbers.

#Median (19): Middle value when numbers are arranged in order.

#Mode (12): Most frequently occurring number.
```

```python
#Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

#Solution:-

import matplotlib.pyplot as plt
import numpy as np


data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data, vert=False)
plt.title("Boxplot of Given Data")
plt.xlabel("Values")
plt.show()


Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("IQR (Interquartile Range):", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)




#Explanation:-

#The boxplot shows the spread of the data through quartiles.

#The box represents the interquartile range (IQR) — where the middle 50% of the data lies.

#The line inside the box is the median (Q2).

#The whiskers extend to the smallest and largest values within 1.5×IQR.

#Points beyond the whiskers are considered outliers.

#Here:
#Q1=17.25, Q3=23.75, IQR=6.5, Upper Bound=33.5, Value 35 > 33.5, so 35 is an outlier.
```

**Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales. **

#Explain how you would use covariance and correlation to explore this relationship.

#Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]   
daily_sales = [2200, 2450, 2750, 3200, 4000]

**Answer**:-
#Covariance

**Definition**: Covariance measures how two variables change together.

1)If both increase together → positive covariance

2)If one increases while the other decreases → negative covariance

However, covariance alone doesn't show how strong the relationship is (it depends on the scale of the data).

#Correlation

**Definition**: Correlation measures the strength and direction of the relationship between two variables.

It's the standardized version of covariance, always between -1 and +1:

+1 → Perfect positive relationship

-1 → Perfect negative relationship

0 → No linear relationship

Compute covariance to check if advertising spend and sales move in the same direction.

Compute correlation to measure how strongly advertising affects sales.

```python
#Program:-

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

cov_matrix = np.cov(advertising_spend, daily_sales)
covariance = cov_matrix[0, 1]

correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Advertising Spend:", advertising_spend)
print("Daily Sales:", daily_sales)
print("\nCovariance:", covariance)
print("Correlation Coefficient:", correlation)
```

#Interpretation

**Covariance** = 78,750 → Positive value ⇒ When ad spend increases, sales also increase.

**Correlation** = 0.9914 → Very close to +1, indicating a strong positive linear relationship.

**Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.**

#Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you'd use.

#Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]  


**Answer**:-To analyze this, you will use 'summary statistics" and "visualizations".

**Summary Statistics:**

-These help describe the center, spread, and shape of the data.

**Visualization**

Histogram — the best tool here — shows how many customers gave each score (1-10).

1)Peaked histogram → most people gave similar ratings.

2)Flat or spread histogram → opinions are varied.

3)Skewed → more low or high ratings.

```python
#Program:-

import numpy as np
import matplotlib.pyplot as plt
import statistics as stats

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)
mode_score = stats.mode(survey_scores)
std_dev = np.std(survey_scores)

print("Mean:", mean_score)
print("Median:", median_score)
print("Mode:", mode_score)
print("Standard Deviation:", std_dev)


plt.hist(survey_scores, bins=6, color='skyblue', edgecolor='black')
plt.title("Customer Satisfaction Survey Distribution")
plt.xlabel("Satisfaction Score (1–10)")
plt.ylabel("Number of Customers")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
